```python
import pandas as pd
import json
```

```python
google_countries = pd.read_csv('countries.csv',encoding='utf-8')
```

```python
google_names = google_countries.values[:,3]
```

```python
es_data = {}
with open('esCountries.json') as json_file:  
    es_data = json.load(json_file)
```

```python
es_names = [i['key'] for i in es_data["aggregations"]["tags"]["buckets"]]
```

```python
not_in_google = []
index = []
for name in es_names:
    try:
        index.append(google_names.tolist().index(name))
    except:
        not_in_google.append(name)
```

```python
def printer():
    for i in not_in_google:
        print(" -> " + i )
printer()
```
